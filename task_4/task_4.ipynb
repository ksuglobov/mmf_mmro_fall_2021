### Вычисление bias и variance с помощью бутстрапа
На лекции была выведено следующая формула, показывающая, как можно представить ошибку алгоритма регрессии в виде суммы трех компонент:
$$
L(\mu) = 
    \mathbb{E}_{x, y}\bigl[\mathbb{E}_{X}\bigl[ (y - \mu(X)(x))^2 \bigr]\bigr] = 
$$
$$
    \underbrace{\mathbb{E}_{x, y}\bigl[(y - \mathbb{E}[y|x] )^2\bigr]}_{\text{шум}} + \underbrace{\mathbb{E}_{x}\bigl[(\mathbb{E}_{X}[\mu(X)(x)] - \mathbb{E}[y|x] )^2\bigr]}_{\text{смещение}} +
    \underbrace{\mathbb{E}_{x}\bigl[\mathbb{E}_{X}\bigl[(\mu(X)(x) - \mathbb{E}_{X}[\mu(X)(x)] )^2\bigr]\bigr]}_{\text{разброс}},
$$
* $\mu(X)$ — алгоритм, обученный по выборке $X = \{(x_1, y_1), \dots (x_\ell, y_\ell)\}$;
* $\mu(X)(x)$ — ответ алгоритма, обученного по выборке $X$, на объекте $x$;
* $\mathbb{E}_{X}$ — мат. ожидание по всем возможным выборкам;
* $\mathbb{E}_{X}[\mu(X)(x)]$ — "средний" ответ алгоритма, обученного по всем возможным выборкам $X$, на объекте $x$.
    
С помощью этой формулы мы можем анализировать свойства алгоритма обучения модели $\mu$, если зададим вероятностную модель порождения пар $p(x, y)$.

В реальных задачах мы, конечно же, не знаем распределение на парах объект - правильный ответ. Однако у нас есть набор семплов из этого распределения (обучающую выборка), и мы можем использовать его, чтобы оценивать математические ожидания. Для оценки мат. ожиданий по выборкам мы будем пользоваться бутстрэпом - методом генерации "новых" выборок из одной с помощью выбора объектов с возвращением. Разберем несколько шагов на пути к оценке смещения и разброса.

#### Приближенное вычисление интегралов
На занятиях мы разбирали примеры аналитического вычисления смещения и разброса нескольких алгоритмов обучения. Для большинства моделей данных и алгоритмов обучения аналитически рассчитать математические ожидания в формулах не удастся. Однако мат. ожидания можно оценивать приближенно. Чтобы оценить математическое ожидание $\mathbb{E}_{\bar z} f(\bar z)$ функции от многомерной случайной величины $\bar z = (z_1, \dots, z_d)$, $\bar z \sim p(\bar z)$, можно сгенерировать выборку из распределения $p(\bar z)$ и усреднить значение функции на элементах этой выборки:
$$\mathbb{E}_{\bar z} f(z) = \int f(\bar z) p(\bar z) d \bar z \approx \frac 1 m \sum_{i=1}^m f(\bar z_i), \, \bar z_i \sim p(\bar z), i = 1, \dots, m.$$

Например, оценим $\mathbb{E}_z z^2,$ $z \sim \mathcal{N}(\mu=5, \sigma=3)$ (из теории вероятностей мы знаем, что
$\mathbb{E}_z z^2 = \sigma^2 + \mu^2 = 34$):

```python
z = np.random.normal(loc=5, scale=3, size=1000)
(z**2).mean()
```

#### Оценивание $\mathbb{E}_{x, y}$
Оценить мат. ожидания по $x$ и по $x, y$, встречающиеся во всех трех компонентах разложения, несложно, потому что у нас есть выборка объектов из распределения данных $p(x, y)$:
$$ \mathbb{E}_{x} f(x) \approx \frac 1 N \sum_{i=1}^N f(x_i), \quad
\mathbb{E}_{x, y} f(x, y) \approx \frac 1 N \sum_{i=1}^N f(x_i, y_i),$$
где $N$ - число объектов в выборке, $\{(x_i, y_i)\}_{i=1}^N$ - сама выборка. 

#### Оценивание $\mathbb{E}_X$ с помощью бутстрапа
Чтобы оценить мат. ожидание по $X$, нам понадобится выборка из выборок:
$$\mathbb{E}_X f(X) \approx \frac 1 s \sum_{j=1}^s f(X_j),$$
где $X_j$ - $j$-я выборка. Чтобы их получить, мы можем воспользоваться бутстрапом - методом генерации выборок на основе выбора объектов с возвращением. Чтобы составить одну выборку, будем $N$ раз выбирать индекс объекта $i \sim \text{Uniform}(1 \dots N)$ и добавлять $i$-ю пару (объект, целевая переменная) в выборку. В результате в каждой выборке могут появиться повторяющиеся объекты, а какие-то объекты могут вовсе не войти в некоторые выборки.

#### Итоговый алгоритм оценки смещения и разброса алгоритма $a$
1. Сгенерировать $s$ выборок $X_j$ методом бутстрапа.
1. На каждой выборке $X_j$ обучить алгоритм $a_j$.
1. Для каждой выборки $X_j$ определить множество объектов $T_j$, не вошедших в нее (out-of-bag). Вычислить предсказания алгоритма $a_j$ на объектах $T_j$. 

Поскольку у нас есть только один ответ для каждого объекта, мы будем считать шум равным 0, а $\mathbb{E}[y|x]$ равным имеющемуся правильному ответу для объекта $x$. 

Итоговые оценки:
* Смещение: для одного объекта - квадрат разности среднего предсказания и правильного ответа. Среднее предсказание берется только по тем алгоритмам $a_j$, для которых этот объект входил в out-of-bag выборку $T_j$. Для получения общего смещения выполнить усреденение смещений по объектам.
* Разброс: для одного объекта - выборочная дисперсия предсказаний алгоритмов $a_j$, для которых этот объект входил в out-of-bag выборку $T_j$. Для получения общего разброса выполнить усреденение разбросов по объектам.
* Ошибка $L$: усреднить квадраты разностей предсказания и правильного ответа по всем выполненным предсказаниям для всех объектов.

В результате должно получиться, что ошибка приблизительно равна сумме смещения и разброса!

Алгоритм также вкратце описан по [ссылке](https://web.engr.oregonstate.edu/~tgd/classes/534/slides/part9.pdf) (слайды 19-21).

Реализуйте описанный алгоритм. Обратите внимание, что если объект не вошел ни в одну из out-of-bag выборок, учитывать его в вычислении итоговых величин не нужно. Как обычно, разрешается использовать только один цикл - по выборкам (от 0 до num_runs-1).

Введём seed (для воспроизводимости) по умолчанию равный 0.

Итак, из исходной выборки S генерируем с помощью бутстрапа $b$ выборок: $S_1, ..., S_b - $ в каждой из которых нет каких-то объектов исходной выборки. Эти объекты образуют $b$ множеств out-of-bag (oob): $T_1, ..., T_b$.



Пусть на $j = \overline{1,n}$ выборках $x_i~-~oob \Rightarrow$

1. Для смещения нам нужны **средние предсказания** на oob объектах. То есть нужно **просуммировать предсказания** на каждом oob объекте и поделить на число раз, когда этот объект, собственно, out-of-bag:
$$\overline{a}(x_i) = \frac{1}{n}\sum_{j=1}^{n}a_j(x_i)$$

2. Для разброса нам нужна **выборочная дисперсия предсказаний** на oob объектах. В алгоритме по [ссылке](https://web.engr.oregonstate.edu/~tgd/classes/534/slides/part9.pdf) (слайды 19-21) используется несмещённая (исправленная) выборочная дисперсия. И мы тоже будем её использовать. Здесь нам уже для каждого объекта потребуется не только **сумма предсказаний**, но и **сумма квадратов предсказаний**:
$$S^2 = \frac{1}{n-1}\sum_{j=1}^{n}(a_j(x_i)-\overline{a}(x_i))^2 = \frac{1}{n-1}\sum_{j=1}^{n}a_j(x_i)^2 - \frac{1}{n(n-1)}\left(\sum_{j=1}^{n}a_j(x_i)\right)^2$$

3. Для вычисления ошибки нам нужен средний по всем предсказаниям по всем объектам квадрат разности правильного ответа и предсказания (mean означает усреднить вектор значений по всем **объектам**, которые хоть раз были oob):
$$L = \left(\frac{1}{n}\sum_{j=1}^{n}(y_j - a_j)^2\right).mean() = \left(\frac{1}{n}\sum_{j=1}^{n}(y_j^2 - 2y_ja_j + a_j^2)\right).mean() = \left(\frac{1}{n}\sum_{j=1}^{n}a_j^2 + \frac{1}{n}\sum_{j=1}^{n}y_j^2 - \frac{2}{n}\sum_{j=1}^{n}y_ja_j\right).mean() = \left\{y_1(x_i) = ... = y_n\right(x_i) = y(x_i)\} = \left(\overline{a(x_i)^2} + y(x_i)^2 - 2y(x_i)\frac{1}{n}\sum_{j=1}^{n}a_j\right).mean() = \left(\overline{a(x_i)^2} + y(x_i)^2 - 2y(x_i)\overline{a(x_i)}\right).mean()$$

$\Rightarrow$ Пусть $s_1 -$ вектор сумм предсказаний oob-объектов, $s_2 -$ вектор сумм квадратов предсказаний oob-объектов, $y -$ вектор правильных объектов для oob-объектов, $mean() -$ усреднение вектора по всем oob-объектам:
1. $bias = \left(\left(y - \frac{1}{n}s_1\right)^2\right).mean()$
2. $variance = (\frac{1}{n-1}s_2 - \frac{1}{n(n-1)}s_1^2).mean()$
3. $error = \left(\frac{1}{n}s_2 + y^2 - \frac{2}{n}ys_1\right).mean()$

Теперь пройдёмся по всем выборкам, по их out-of-bag объектам, сформируем векторы и посчитаем искомые величины по формулам:

```python
def compute_biase_variance(regressor, X, y, num_runs=1000, seed=0):    
    """
    :param regressor: sklearn estimator with fit(...) and predict(...) method
    :param X: numpy-array representing training set ob objects, shape [n_obj, n_feat]
    :param y: numpy-array representing target for training objects, shape [n_obj]
    :param num_runs: int, number of samples (s in the description of the algorithm)
    
    :returns: bias (float), variance (float), error (float) 
    each value is computed using bootstrap
    """
    y_len = len(y)
    # random generator
    rng = np.random.default_rng(seed=seed)
    # original dataset indexes
    all_ind = np.arange(y_len)

    # cumulative values: oob counts, oob pred sums, oob pred sq sums
    cnt = np.zeros_like(all_ind, dtype=int)
    s1 = np.zeros_like(all_ind, dtype=float)
    s2 = np.zeros_like(all_ind, dtype=float)
    for i in trange(num_runs):
        # bootstrap - index generation
        bootstrap_ind = rng.choice(y_len, y_len)
        # oob dataset indexes
        oob_ind = np.setdiff1d(all_ind, bootstrap_ind)

        # fit-predict
        regressor.fit(X[bootstrap_ind], y[bootstrap_ind])
        y_pred = regressor.predict(X[oob_ind])

        # cumulative values
        cnt[oob_ind] += 1
        s1[oob_ind] += y_pred
        s2[oob_ind] += y_pred ** 2
    # taking only oob
    s1, s2, y, cnt = s1[cnt > 0], s2[cnt > 0], y[cnt > 0], cnt[cnt > 0]
    # results
    bias = ((s1 / cnt - y) ** 2).mean()
    variance = ((s2 - s1 ** 2 / cnt) / (cnt - 1)).mean()
    error = (s2 / cnt + y * (y - 2 * s1 / cnt)).mean()
    return bias, variance, error
```

```python
from sklearn.model_selection import train_test_split
```

```python
def plot_predictions(regressor, X, y, num_runs=100, num_test_objects=10, title=""):
    """
    plot graphics described above
    """
    # generator
    rng = np.random.default_rng(seed=0)
    # test objects indexes
    all_ind = np.arange(num_test_objects)
    # samples
    X_train, X_test, y_train, y_test = train_test_split(X, y,
            test_size=num_test_objects, random_state=0)
    y_train_len = len(y_train)
    for i in range(num_runs):
        bootstrap_ind = rng.choice(y_train_len, y_train_len)
        # fit-predict
        regressor.fit(X_train[bootstrap_ind], y_train[bootstrap_ind])
        y_pred = regressor.predict(X_test)
        #plot
        plt.scatter(all_ind, y_pred, color='salmon', s=75)
    plt.scatter(all_ind, y_test, color='blue', marker='x', s=65)
    plt.title(title)
    plt.xlabel('test objects')
    plt.xticks(all_ind)
```

```python
def nduplicates(df, col, n=5):
    # n largest duplicates (dup values in rows)
    ndups = df.groupby(col).size().sort_values(ascending=False)[:n]
    # n largest duplicates (rows)
    all_dups = df.loc[df[col].isin(ndups.index)]
    # dup rows aggregation
    res = all_dups.groupby(col).agg(lambda vals: set(vals)).reindex(ndups.index)
    # adding col with counts
    res.insert(1, 'Counts', ndups.values)
    return res
```

Функция для быстрого получения ROC-AUC модели:

```python
def get_auc(clf, X_train, y_train, X_test, y_test):
    clf.fit(X_train, y_train)
    y_proba = clf.predict_proba(X_test)[:, 1]
    return pr_auc(y_test, y_proba)
```

```python
# encoding
ohe_encoder = ce.one_hot.OneHotEncoder(cols=str_feats)
# fitting
start_time = time.time()
ohe_encoder.fit(X_train)
print('OHE fit time: ', time.time() - start_time, 's')
#transforming
X_train_ohe, X_test_ohe = (ohe_encoder.transform(X_train),
                           ohe_encoder.transform(X_test))

# encoding
smoothed_cnt_encoder = (ce.target_encoder
                        .TargetEncoder(cols=str_feats, smoothing=1))
# fitting
start_time = time.time()
smoothed_cnt_encoder.fit(X_train, y_train)
print('Smoothed counters fit time: ', time.time() - start_time, 's')
# transforming
X_train_cnt, X_test_cnt = (smoothed_cnt_encoder.transform(X_train),
                           smoothed_cnt_encoder.transform(X_test))

enc_arr = ['no categorical', 'one-hot', 'smoothed counts']
enc_train_arr, enc_test_arr = ([X_train[num_feats], X_train_ohe, X_train_cnt],
                               [X_test[num_feats], X_test_ohe, X_test_cnt])
```

Сравним работу LightGBM и Catboost на разных способах кодирования (и без кодирования) на подобраных в предыдущем эксперименте лучших гиперпараметрах:

```python
lightgbm_params = {
    'random_state': 0,
    'max_depth': 4,
    'n_estimators': 400,
    'learning_rate': 0.05,  
}

catboost_params = {
    'random_state': 0,
    'verbose': 0,
    'max_depth': 4,
    'n_estimators': 600,
    'learning_rate': 0.1,
}

clf_arr = {
    'LightGBM': LGBMClassifier(**lightgbm_params),
    'CatBoost': CatBoostClassifier(**catboost_params),
}
```

Запустим каждую комбинацию параметров 3 раза, время усредним:

```python
res_ind = pd.MultiIndex.from_product([clf_arr.keys(), ['time', 'PR-AUC']])
res = pd.DataFrame(index=res_ind, columns=enc_arr)
```

```python
for clf_ in clf_arr:
    for enc_, Xtr, Xts in zip(enc_arr, enc_train_arr, enc_test_arr):
        time_ = 0
        for j in range(3):
            start_time = time.time()
            clf_arr[clf_].fit(Xtr, y_train)
            time_ += time.time() - start_time
        time_ /= 3
        # writing time in res
        res.loc[(clf_, 'time'), enc_] = round(time_, 2)
        # auc
        auc = pr_auc(y_test, clf_arr[clf_].predict_proba(Xts)[:, 1])
        # writing auc in res
        res.loc[(clf_, 'PR-AUC'), enc_] = auc
```

При этом эта разница во времени обучения несущественна из-за небольшого признакового пространства в градиентном бустинге даже при one-hot кодировании (это максимальное число признаков)
